head(10) vs nrows=10

* head() loads entire dataframe into the memory and then display the 10 rows. 
* Memory intensive operation 
* This will take much time to display

* whereas nrows will just fetch 10 rows, loads into the memory and then display
* Doesnt consume much time as head()
